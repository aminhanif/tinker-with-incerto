```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
var_list = np.array([])
std_list = np.array([])
mean_list = np.array([])
std_of_mean = np.array([])

gaussian = np.random.randn(25000)
print(np.std(gaussian))

for variable in gaussian:
    var_list = np.append(var_list, variable)
    mean = np.mean(var_list)
    mean_list = np.append(mean_list, mean)
    std_list = np.append(std_list, np.std(var_list))
    std_of_mean = np.append(std_of_mean, np.std(mean_list))

x_coor = np.arange(1, len(gaussian)+1)
plt.plot(x_coor, std_list)
plt.show()

plt.plot(x_coor, mean_list)
plt.show()
```

Look how the STD and STD of mean converge and stabilize as the number of variables added!!!

```python
import numpy as np
import matplotlib.pyplot as plt

alpha = 2
size = 10**3

pareto_samples = (np.random.pareto(alpha, size) + 1)
#pareto_samples *= (10000 / np.max(pareto_samples))

print(np.max(pareto_samples))


plt.figure(figsize=(8,6))
plt.hist(pareto_samples, bins=400, density=True, alpha=1)

x = np.linspace(1, 10, 1000)
pdf = alpha * x**(-(alpha + 1))

plt.plot(x, pdf, 'r-', lw=2, label=f'Theoretical PDF (alpha = {alpha})')

plt.title('Pareto distribution')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.grid()
plt.show()
```

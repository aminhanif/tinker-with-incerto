```python
# THE GOAL IS TO DEMONSTRATE HOW STD (STANDARD DEVIATION) AND MAD (MEAN ABSOLUTE DEVIATION) YIELD DIFF. RESULTS IN FAT TAILED DATA

from scipy.stats import norm
import matplotlib.pyplot as plt
import numpy as np
from numpy import random
```

```python
x_g = np.random.randn(5000)
mean = np.mean(x_g)
std = np.std(x_g)
mad = np.mean(np.abs(x_g - mean))
print(mean, std, mad, std/mad)
plt.hist(x_g, bins=300, color='green')
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram of X")
```

```python
x = np.append(np.zeros(999999), [10**6])
mean = np.mean(x)
std = np.std(x)
mad = np.mean(np.abs(x - mean))
print(mean, std, mad, std/mad)
plt.hist(x, bins=30, color='green')
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram of X")
```

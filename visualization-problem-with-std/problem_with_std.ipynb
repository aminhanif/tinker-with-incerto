```python
# THE GOAL IS TO DEMONSTRATE HOW STD (STANDARD DEVIATION) AND MAD (MEAN ABSOLUTE DEVIATION) YIELD DIFF. RESULTS IN FAT TAILED DATA

from scipy.stats import norm
import matplotlib.pyplot as plt
import numpy as np
from numpy import random
```

## STD AND MAD IN GAUSSIAN DISTRIBUTION

```python
x_g = np.random.randn(5000)
plt.hist(x_g, bins=300, color='green')
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram of X")
mean = np.mean(x_g).round(2)
std = np.std(x_g).round(2)
mad = np.mean(np.abs(x_g - mean)).round(2)
print(f"mean = {mean} | std = {std} | mad = {mad}\nstd/mad = {std/mad}")
```

## STD AND MAD IN VERY THICK-TAILED DISTRIBUTION

now lets take a look at distribution of a million variables consist of 999999 zeros and and a 10^6 --> [0, 0, 0 .... 0, 10^6]

```python
x = np.append(np.zeros(999999), [10**6])
mean = np.mean(x).round(2)
std = np.std(x).round(2)
mad = np.mean(np.abs(x - mean)).round(2)
print(f"mean = {mean} | std = {std} | mad = {mad}\nstd/mad = {std/mad}")
plt.hist(x, bins=10, color='green')
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram of X")
```

## CONCLUSION

1. In gaussian distribution, STD and MAD closely resembling each other. While in fat tailed distribution, there are huge different between the two (while STD can be very very large). 

2. Conclusion #1 happened because the quadratic formula of STD will disproportionately make the big number even bigger, things that is not happen in MAD. 

3. Hence, in gaussian settings, you can assume that STD is the same as MAD (their difference is only 1.25). But in fat-tailed distribution you can't.

4. STD alone contain very little informational value in fat tail distribution. Consider in Figure 2 that the STD are 1000. Intuitively, we consider from that value that in general the data points will have a 1000 different in values than the mean (which is 1). So we expect the data points to be around 1001, 1002, 1008, etc. But in truth ALMOST ALL OF THEM WERE 0 (only 1 different in values than the mean)!!!. Basically, the one data points (10^6) just change everything. 

5. In Figure 2, the MAD (with value 2) actually have more resemblance for most of the data (where 999.999 of them are 0).

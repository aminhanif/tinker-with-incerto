```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
import seaborn as sns
```

```python
# Load your data (CSV file) into a DataFrame
df = pd.read_csv('height.csv')

# Choose the column you want to plot
data = df['Height(Inches)'] * 2.54 # Replace with the name of your column, make it in cm
```

```python
# Calculate mean and standard deviation
mean = np.mean(data)
std_dev = np.std(data)

# Set Seaborn style for better aesthetics
sns.set(style="whitegrid")

# Plot histogram with KDE
plt.figure(figsize=(10, 6))
sns.histplot(data, bins=30, kde=True, color='blue', alpha=0.6)

# Add mean and standard deviation lines
plt.axvline(mean, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {mean:.0f}')
plt.axvline(mean + std_dev, color='green', linestyle='dotted', linewidth=2, label=f'+1 Std Dev: {mean + std_dev:.0f}')
plt.axvline(mean - std_dev, color='green', linestyle='dotted', linewidth=2, label=f'-1 Std Dev: {mean - std_dev:.0f}')
plt.axvline(mean + 2*std_dev, color='green', linestyle='dotted', linewidth=2, label=f'+2 Std Dev: {mean + 2*std_dev:.0f}')
plt.axvline(mean + 3*std_dev, color='green', linestyle='dotted', linewidth=2, label=f'+3 Std Dev: {mean + 3*std_dev:.0f}')
plt.axvline(mean + 4*std_dev, color='green', linestyle='dotted', linewidth=2, label=f'+4 Std Dev: {mean + 4*std_dev:.0f}')
plt.axvline(mean + 5*std_dev, color='green', linestyle='dotted', linewidth=2, label=f'+5 Std Dev: {mean + 5*std_dev:.0f}')
plt.axvline(mean + 6*std_dev, color='green', linestyle='dotted', linewidth=2, label=f'+6 Std Dev: {mean + 6*std_dev:.0f}')
plt.axvline(mean + 9*std_dev, color='green', linestyle='dotted', linewidth=2, label=f'+9 Std Dev: {mean + 9*std_dev:.0f}')


# Add labels, title, and legend
plt.title(f'Distribution of Data\nMean = {mean:.0f}, Std Dev = {std_dev:.0f}')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()

# Show the plot
plt.show()
```

```python
in_2K = mean + 3*std_dev
in_Ks = mean + 9*std_dev
prob_2K = (len(data) - len(data[(data < in_2K)])) / len(data)
prob_Ks = (len(data) - len(data[(data < in_Ks)])) / len(data)
print(prob_Ks/(prob_2K*prob_2K))
```

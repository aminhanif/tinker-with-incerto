```python
# TRY TO VISUALIZE VARIOUS REAL-WORLD DATA AND SEE SOME PROPERTIES: 
# 1. PERCENTAGE OF DATA THAT FALL WITHIN STD, 2 STD, 3 STD, AND SO ON
# 2. MEAN, VARIANCE, KURTOSIS, AND SO ON
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## GAUSSIAN RANDOM NUMBER

```python
gaussian = np.random.randn(25000)
values = gaussian
mean = np.mean(values).round(2)
std = np.std(values).round(2)
plt.hist(gaussian, bins=300, color='green')
plt.axvline(mean + std, color ='blue', label='1 STD')
plt.axvline(mean - std, color ='blue', label='-1 STD')
plt.axvline(mean + (2 * std), color ='yellow', label='2 STD')
plt.axvline(mean - (2 * std), color ='yellow', label='-2 STD')
plt.axvline(mean + (3 * std), color ='purple', label='3 STD')
plt.axvline(mean - (3 * std), color ='purple', label='-3 STD')
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram of X")

total_data = len(values)
within_1_std = (((values <= mean + std) & (values >= mean-std)).sum()/total_data*100).round(2)
within_2_std = (((values <= mean + (2*std)) & (values >= mean - (2*std))).sum()/total_data*100).round(2)
within_3_std = (((values <= mean + (3*std)) & (values >= mean - (3*std))).sum()/total_data*100).round(2)
within_4_std = (((values <= mean + (4*std)) & (values >= mean - (4*std))).sum()/total_data*100).round(2)
within_5_std = (((values <= mean + (5*std)) & (values >= mean - (5*std))).sum()/total_data*100).round(2)

plt.text(-6, -80, f"mean = {mean} | std = {std}")
plt.text(-6, -100, f"within 1 STD = {within_1_std}% | within 2 STD = {within_2_std}% | within 3 STD = {within_3_std}% | within 4 STD = {within_4_std}% | within 5 STD = {within_5_std}%")
```

## ALL DATA VISUALIZATION

- (In gaussian) the probability of an event staying within one standard deviation of the mean is 68 percent. As the tails fatten, to mimic what happens in financial market for example, the probability of an event staying within one standard deviation of the mean rises between 75 and 95 percent. So note that **as we fatten the tails we get higher peaks, smaller shoulders, and a higher incidence of a very large deviation**. Because probabilities need to add up to 1 (even in France) increasing mass in one area leads to decreasing it in another - Technical Incerto, p. 23 

- Now, notice what happens. Figure 1 and 2 is gaussian (about 68 percent of data fall within 1 std and nothing more than 5 std). Notice how Figure 3 and 4 is very unlike them both. A REALLY HIGH percentage of data (>90%) fall within 1 STD BUT there is data that fall more than 5 STD!!! 

- Also notice how from within 3 STD onward the percentage data that fall within it is lesser than gaussian. Hence visually you can see that it is fat tailed: higher peaks, smaller shoulders and the incidence of a very large deviation.

- From this, notice the indication that variable from nature (height) resemble gaussian distribution and the "human made" variable (country population and gold value) is fat tailed
